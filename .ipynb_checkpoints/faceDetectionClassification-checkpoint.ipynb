```python
#### Functions for Part 1
import cv2
import matplotlib.pyplot as plt
import numpy as np
import copy
import math
import glob



def resize(img):
    #Takes an input image and returns the same image in correct size
    #and in black and white
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img =cv2.resize(img, (150, 200))
    height, width = img.shape
    return img

def showImg(img, title):
    plt.imshow(img, cmap='gray'), plt.title(str(title))
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def showImgC(img, title):
    plt.imshow(img), plt.title(str(title))
    plt.xticks([]), plt.yticks([])
    plt.show()

#Functions for Part 2

def blurImage(img_gray):
    #Takes an image and returns the smoothed version of said image
    height, width = img_gray.shape
    pheight = height + 14
    pwidth = width + 14
    ar2 = np.zeros((height, width), np.float)
    ar = np.zeros((pheight, pwidth), np.float)
    ar[7:height+7, 7:width+7] = img_gray
    size = 15
    avefilt = np.ones((size, size), np.float)/(size * size)
    for row in range(0, height):
        for column in range(0, width):
            ar2[row, column] = (ar[row:(row+size), column:(column+size)]*avefilt).sum()
    return ar2


def openBlurSob(img):
    #cv2.smooth(img, img, smoothtype = CV_BLUR, param1 = 3, param2=3)
    img = cv2.Sobel(img,cv2.CV_64F,1,1,ksize=5)
    return img

def makeBinary(img, thres):
    #Take the sobel img and convert it to binary
    height, width = img.shape
    for row in range(0, height):
        for column in range(0, width):
            if(img[row, column] >= thres):
                img[row, column] = 0
            else:
                img[row, column] = 255
    return img

def draw_boundary(img2, w1, w3, upper, lower, width, height):
    cv2.line(img2, (0, upper), (width, upper), (255, 0 , 0), 1)
    cv2.line(img2, (0, lower), (width, lower), (255, 0, 0), 1)
    cv2.line(img2, (w1, 0), (w1, height), (255, 0, 0), 1)
    cv2.line(img2, (w3, 0), (w3, height), (255, 0, 0), 1)
    return img2

def eyeSearch(img, img2):
    height, width = img.shape
    upper = int(height/3)
    lower = int((3*height)/5)
    hist = np.array([])
    w1 = int(width/4)
    w3 = int(3*width/4)
    
    for row in range(upper, lower):
        sum_intensity = 0
        intensityp_sum = 0
        for width in range(w1, w3):
            sum_intensity = img[row][width] + sum_intensity
            intensityp_sum =(img[row-1][width] + intensityp_sum)
        intensityp = int(intensityp_sum)
        intensity = int(sum_intensity)
        if row == upper:
            #intensity_delta = intensity
            intensity_delta = int(intensity - intensityp)
            hist = np.append(hist, intensity_delta)
            strongest_row = row
            strongest_delta = intensity_delta
        else:
            #intensity_delta = intensity
            intensity_delta = int(intensity - intensityp)
            hist = np.append(hist, intensity_delta)
            if intensity_delta > strongest_delta:
                strongest_row = row
                strongest_delta = intensity_delta
        #intensityp = intensity
    print(strongest_row)
    draw_boundary(img2, w1, w3, upper, lower, width, height)
    cv2.line(img2, (0, strongest_row), (width, strongest_row), (255, 0 , 0), 1)
    showImg(img2, "Eye Bar")
    

                
def main():
    image_path = "/media/dalton/Data/ECSE415/Project/Final_Project_Faces/000011.jpg"
    img = cv2.imread(image_path)
    img = resize(img)
    height, width = img.shape
    img2 = img.copy()
    showImg(img, "Resized Gray")
    
    img = blurImage(img)
    #img = openBlurSob(img)
    showImg(img, "blurred")
    
    #img = sobel(img)
    img = openBlurSob(img)
    showImg(img, "sobel")
    
    img = makeBinary(img, 30)
    showImg(img, "Binary")
    
    eye_row = eyeSearch(img, img2)
    

main()
```

```python
#### Functions for Part 1
import cv2
import matplotlib.pyplot as plt
import numpy as np
import copy
import math
import glob
import random
import os
import time

t0 = time.time()
def showImg(img, title):
    plt.imshow(img, cmap='gray'), plt.title(str(title))
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def showImgC(img, title):
    plt.imshow(img), plt.title(str(title))
    plt.xticks([]), plt.yticks([])
    plt.show()

data_age_path = '/home/dalton/ECSE415/Project/Ages_new.txt'
with open(data_age_path) as f:
    age = f.readlines()
age = [line.rstrip('\n') for line in open(data_age_path)]

#paths to files
xml_path = '/home/dalton/anaconda3/share/OpenCV/haarcascades/'
faces_path = '/home/dalton/ECSE415/Project/Final_Project_Faces/'


#creating haar classifiers from xml files
face_cascade = cv2.CascadeClassifier(xml_path + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(xml_path + 'haarcascade_eye.xml')
mouth_cascade = cv2.CascadeClassifier(xml_path + 'haarcascade_mouth.xml')
nose_cascade = cv2.CascadeClassifier(xml_path + 'haarcascade_nose.xml')


#lists to stores training faces
faces_list = []
gray_faces_list = []

#loading training faces
files = os.listdir(faces_path)
files.sort()
for filename in files:
    img=cv2.imread(faces_path+filename)
    img = cv2.resize(img,(150, 200))
    faces_list.append(img)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_faces_list.append(gray)

t1 = time.time()
print("Time used",t1-t0)
```

```python
# Part 2 Stage 1

#copying faces ready to be drawn on

def VJadaboost(faces_list, faces_draw, gray_faces_list, text):
    #initialising feature location storage arrays
    eye_right_height = np.zeros((len(faces_list),1))
    eye_left_height = np.zeros((len(faces_list),1))

    eye_right_upper = np.zeros((len(faces_list),1))
    eye_right_lower = np.zeros((len(faces_list),1))
    eye_right_outer = np.zeros((len(faces_list),1))
    eye_right_inner = np.zeros((len(faces_list),1))
    eye_left_upper = np.zeros((len(faces_list),1))
    eye_left_lower = np.zeros((len(faces_list),1))
    eye_left_outer = np.zeros((len(faces_list),1))
    eye_left_inner = np.zeros((len(faces_list),1))

    eye_ave_upper = np.zeros((len(faces_list),1))
    eye_ave_height = np.zeros((len(faces_list),1))
    eye_ave_lower = np.zeros((len(faces_list),1))

    mouth_height = np.zeros((len(faces_list),1))

    mouth_upper = np.zeros((len(faces_list),1))
    mouth_lower = np.zeros((len(faces_list),1))
    mouth_left = np.zeros((len(faces_list),1))
    mouth_right = np.zeros((len(faces_list),1))

    nose_height = np.zeros((len(faces_list),1))

    nose_upper = np.zeros((len(faces_list),1))
    nose_lower = np.zeros((len(faces_list),1))
    nose_left = np.zeros((len(faces_list),1))
    nose_right = np.zeros((len(faces_list),1))

    #iterating through faces
    for i in range(len(faces_list)):
        #calling face finder method
        faces = face_cascade.detectMultiScale(gray_faces_list[i], 1.3, 5)
        for (x,y,w,h) in faces:
            #drawing bounding rectangle on face
            #cv2.rectangle(faces_draw[i],(x,y),(x+w,y+h),(255,0,0),2)
            #creating smaller error in which to search for facial features
            roi_gray = gray_faces_list[i][y:y+h, x:x+w]
            roi_color = faces_draw[i][y:y+h, x:x+w]
            #finding facial features using haar cascades
            eyes = eye_cascade.detectMultiScale(roi_gray)
            nose = nose_cascade.detectMultiScale(roi_gray)
            eye_no = 1
            for (ex,ey,ew,eh) in eyes:
                #identifying and storing eye bounding lines
                if eye_no == 1:
                    temp_ex = ex
                    temp_ey = ey
                    temp_ew = ew
                    temp_eh = eh
                    #drawing bounding rectangle round eyes
                    cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

                elif eye_no == 2:
                    if ex>temp_ex:
                        eye_right_height[i] = int(y+ey+(0.5*eh))
                        eye_right_upper[i] = y+ey
                        eye_right_lower[i] = y+ey+eh
                        eye_right_outer[i] = x+ex+ew
                        eye_right_inner[i] = x+ex
                        eye_left_height[i] = int(y+temp_ey+(0.5*temp_eh))
                        eye_left_upper[i] = y+temp_ey
                        eye_left_lower[i] = y+temp_ey+temp_eh
                        eye_left_outer[i] = x+temp_ex
                        eye_left_inner[i] = x+temp_ex+temp_ew
                    else:
                        eye_right_height[i] = int(y+temp_ey+(0.5*temp_eh))
                        eye_right_upper[i] = y+temp_ey
                        eye_right_lower[i] = y+temp_ey+temp_eh
                        eye_right_outer[i] = x+temp_ex+temp_ew
                        eye_right_inner[i] = x+temp_ex
                        eye_left_height[i] = int(y+ey+(0.5*eh))
                        eye_left_upper[i] = y+ey
                        eye_left_lower[i] = y+ey+eh
                        eye_left_outer[i] = x+ex
                        eye_left_inner[i] = x+ex+ew

                        eye_ave_upper[i] = int(0.5*(eye_right_upper[i]+eye_left_upper[i]))
                        eye_ave_height[i] = int(0.5*(eye_right_height[i]+eye_left_height[i]))
                        eye_ave_lower[i] = int(0.5*(eye_right_lower[i]+eye_left_lower[i]))
                        #drawing bounding rectangle round eyes
                        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
                        cv2.line(faces_draw[i],(0,eye_ave_height[i]),(faces_draw[i].shape[0],eye_ave_height[i]),(0,255,0),1)

                eye_no += 1

            nose_no = 1
            for (nx,ny,nw,nh) in nose:
                if nose_no == 1:
                    #finding nose and storing location
                    nose_x = int(0.5*(x+nx+x+nx+nw))
                    cv2.line(faces_draw[i],(nose_x,0),(nose_x,faces_draw[i].shape[1]),(0,0,255),1)
                    nose_height[i] = int(y+ny+(0.5*nh))
                    nose_upper[i] = y+ny
                    nose_lower[i] = y+ny+nh
                    nose_right[i] = x+nx+nw
                    nose_left[i] = x+nx
                    nose_no += 1

            roi_gray = gray_faces_list[i][y+int(0.6*h):y+h, x:x+w]
            roi_color = faces_draw[i][y+int(0.6*h):y+h, x:x+w]
            #roi_gray = gray_faces_list[i][y:y+int(0.45*h), x:x+w]
            #roi_color = faces_draw[i][y:y+int(0.45*h), x:x+w]
            mouth = mouth_cascade.detectMultiScale(roi_gray)
            mouth_no = 1
            for (mx,my,mw,mh) in mouth:
                if mouth_no == 1:
                    #finding mouth location and storing location data
                    cv2.rectangle(roi_color,(mx,my),(mx+mw,my+mh),(80,150,0),2)
                    mouth_upper[i] = y+int(0.5*h)+my
                    mouth_lower[i] = y+int(0.5*h)+my+mh
                    mouth_left[i] = x+mx
                    mouth_right[i] = x+mx+mw
                    mouth_height[i] = int(y+(0.6*h)+my+(0.5*mh))
                    cv2.line(faces_draw[i],(0,mouth_height[i]),(faces_draw[i].shape[0],mouth_height[i]),(80,150,0),1)
                    mouth_no += 1

    sample = random.sample(range(len(faces_list)), 5)
    for i in sample:
        faces_draw[i] = cv2.cvtColor(faces_draw[i], cv2.COLOR_BGR2RGB)
        showImg(faces_draw[i],text)
        
VJadaboost(babyC, babyC.copy(),baby.copy(),"Baby face with highlighted features.")
VJadaboost(youngAdultsC, youngAdultsC.copy(),youngAdults.copy(),'Young adults face with highlighted features.')
VJadaboost(middleAgedAdultsC, middleAgedAdultsC.copy(),middleAgedAdults.copy(),'Middle aged adults face with features.')
VJadaboost(oldAdultsC, oldAdultsC.copy(), oldAdults.copy(),'Old Adults face with features.')
```

```python
t2=time.time()
def featurecompute(faces_list, faces_draw, gray_faces_list,):
    #initialising feature location storage arrays
    eye_right_height = np.zeros((len(faces_list),1))
    eye_left_height = np.zeros((len(faces_list),1))

    eye_right_upper = np.zeros((len(faces_list),1))
    eye_right_lower = np.zeros((len(faces_list),1))
    eye_right_outer = np.zeros((len(faces_list),1))
    eye_right_inner = np.zeros((len(faces_list),1))
    eye_left_upper = np.zeros((len(faces_list),1))
    eye_left_lower = np.zeros((len(faces_list),1))
    eye_left_outer = np.zeros((len(faces_list),1))
    eye_left_inner = np.zeros((len(faces_list),1))

    eye_ave_upper = np.zeros((len(faces_list),1))
    eye_ave_height = np.zeros((len(faces_list),1))
    eye_ave_lower = np.zeros((len(faces_list),1))

    mouth_height = np.zeros((len(faces_list),1))

    mouth_upper = np.zeros((len(faces_list),1))
    mouth_lower = np.zeros((len(faces_list),1))
    mouth_left = np.zeros((len(faces_list),1))
    mouth_right = np.zeros((len(faces_list),1))

    nose_height = np.zeros((len(faces_list),1))

    nose_upper = np.zeros((len(faces_list),1))
    nose_lower = np.zeros((len(faces_list),1))
    nose_left = np.zeros((len(faces_list),1))
    nose_right = np.zeros((len(faces_list),1))

    #iterating through faces
    for i in range(len(faces_list)):
        #calling face finder method
        faces = face_cascade.detectMultiScale(gray_faces_list[i], 1.3, 5)
        for (x,y,w,h) in faces:
            #drawing bounding rectangle on face
            #cv2.rectangle(faces_draw[i],(x,y),(x+w,y+h),(255,0,0),2)
            #creating smaller error in which to search for facial features
            roi_gray = gray_faces_list[i][y:y+h, x:x+w]
            roi_color = faces_draw[i][y:y+h, x:x+w]
            #finding facial features using haar cascades
            eyes = eye_cascade.detectMultiScale(roi_gray)
            nose = nose_cascade.detectMultiScale(roi_gray)
            eye_no = 1
            for (ex,ey,ew,eh) in eyes:
                #identifying and storing eye bounding lines
                if eye_no == 1:
                    temp_ex = ex
                    temp_ey = ey
                    temp_ew = ew
                    temp_eh = eh
                    #drawing bounding rectangle round eyes
                    cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

                elif eye_no == 2:
                    if ex>temp_ex:
                        eye_right_height[i] = int(y+ey+(0.5*eh))
                        eye_right_upper[i] = y+ey
                        eye_right_lower[i] = y+ey+eh
                        eye_right_outer[i] = x+ex+ew
                        eye_right_inner[i] = x+ex
                        eye_left_height[i] = int(y+temp_ey+(0.5*temp_eh))
                        eye_left_upper[i] = y+temp_ey
                        eye_left_lower[i] = y+temp_ey+temp_eh
                        eye_left_outer[i] = x+temp_ex
                        eye_left_inner[i] = x+temp_ex+temp_ew
                    else:
                        eye_right_height[i] = int(y+temp_ey+(0.5*temp_eh))
                        eye_right_upper[i] = y+temp_ey
                        eye_right_lower[i] = y+temp_ey+temp_eh
                        eye_right_outer[i] = x+temp_ex+temp_ew
                        eye_right_inner[i] = x+temp_ex
                        eye_left_height[i] = int(y+ey+(0.5*eh))
                        eye_left_upper[i] = y+ey
                        eye_left_lower[i] = y+ey+eh
                        eye_left_outer[i] = x+ex
                        eye_left_inner[i] = x+ex+ew

                        eye_ave_upper[i] = int(0.5*(eye_right_upper[i]+eye_left_upper[i]))
                        eye_ave_height[i] = int(0.5*(eye_right_height[i]+eye_left_height[i]))
                        eye_ave_lower[i] = int(0.5*(eye_right_lower[i]+eye_left_lower[i]))
                        #drawing bounding rectangle round eyes
                        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
                        cv2.line(faces_draw[i],(0,eye_ave_height[i]),(faces_draw[i].shape[0],eye_ave_height[i]),(0,255,0),1)

                eye_no += 1

            nose_no = 1
            for (nx,ny,nw,nh) in nose:
                if nose_no == 1:
                    #finding nose and storing location
                    nose_x = int(0.5*(x+nx+x+nx+nw))
                    cv2.line(faces_draw[i],(nose_x,0),(nose_x,faces_draw[i].shape[1]),(0,0,255),1)
                    nose_height[i] = int(y+ny+(0.5*nh))
                    nose_upper[i] = y+ny
                    nose_lower[i] = y+ny+nh
                    nose_right[i] = x+nx+nw
                    nose_left[i] = x+nx
                    nose_no += 1

            roi_gray = gray_faces_list[i][y+int(0.6*h):y+h, x:x+w]
            roi_color = faces_draw[i][y+int(0.6*h):y+h, x:x+w]
            #roi_gray = gray_faces_list[i][y:y+int(0.45*h), x:x+w]
            #roi_color = faces_draw[i][y:y+int(0.45*h), x:x+w]
            mouth = mouth_cascade.detectMultiScale(roi_gray)
            mouth_no = 1
            for (mx,my,mw,mh) in mouth:
                if mouth_no == 1:
                    #finding mouth location and storing location data
                    cv2.rectangle(roi_color,(mx,my),(mx+mw,my+mh),(80,150,0),2)
                    mouth_upper[i] = y+int(0.5*h)+my
                    mouth_lower[i] = y+int(0.5*h)+my+mh
                    mouth_left[i] = x+mx
                    mouth_right[i] = x+mx+mw
                    mouth_height[i] = int(y+(0.6*h)+my+(0.5*mh))
                    cv2.line(faces_draw[i],(0,mouth_height[i]),(faces_draw[i].shape[0],mouth_height[i]),(80,150,0),1)
                    mouth_no += 1

    
    
    cheek_right_density = np.zeros((len(faces_list),1),np.float32)
    cheek_left_density = np.zeros((len(faces_list),1),np.float32)
    eye_corners_right_density = np.zeros((len(faces_list),1),np.float32)
    eye_corners_left_density = np.zeros((len(faces_list),1),np.float32)
    brow_density = np.zeros((len(faces_list),1),np.float32)

    cheek_right_depth = np.zeros((len(faces_list),1),np.float32)
    cheek_left_depth = np.zeros((len(faces_list),1),np.float32)
    eye_corners_right_depth = np.zeros((len(faces_list),1),np.float32)
    eye_corners_left_depth = np.zeros((len(faces_list),1),np.float32)
    brow_depth = np.zeros((len(faces_list),1),np.float32)

    cheek_right_var = np.zeros((len(faces_list),1),np.float32)
    cheek_left_var = np.zeros((len(faces_list),1),np.float32)
    eye_corners_right_var = np.zeros((len(faces_list),1),np.float32)
    eye_corners_left_var = np.zeros((len(faces_list),1),np.float32)
    brow_var = np.zeros((len(faces_list),1),np.float32)

    R_em = np.zeros((len(faces_list),1),np.float32)
    R_enm = np.zeros((len(faces_list),1),np.float32)

    face_features = np.zeros((len(faces_list),17),np.float32)

    cheeks_right_lower = (mouth_height).astype(int)
    cheeks_right_upper = (eye_right_lower).astype(int)
    cheeks_right_inner = (nose_right).astype(int)
    cheeks_right_outer = (eye_right_outer).astype(int)
    cheeks_left_lower = mouth_height
    cheeks_left_upper = eye_left_lower
    cheeks_left_inner = nose_left
    cheeks_left_outer = eye_left_outer

    eye_corners_right_upper = eye_right_upper
    eye_corners_right_lower = eye_right_lower
    eye_corners_right_inner = eye_right_outer
    eye_corners_right_outer = ((0.3*(eye_right_outer-eye_right_inner))+eye_right_outer).astype(int)
    eye_corners_left_upper = eye_left_upper
    eye_corners_left_lower = eye_left_lower
    eye_corners_left_inner = eye_left_outer
    eye_corners_left_outer = (eye_left_outer-(0.3*(eye_left_inner-eye_left_outer))).astype(int)

    brow_lower = eye_ave_upper
    brow_upper = ((0.5*(eye_ave_upper-eye_ave_lower))+brow_lower).astype(int)
    brow_right = ((eye_right_outer+eye_right_inner)*0.5).astype(int)
    brow_left = ((eye_left_inner+eye_left_outer)*0.5).astype(int)

    D_em = eye_ave_height-mouth_height
    D_ee = ((0.5*(eye_right_outer+eye_right_inner))-(0.5*(eye_left_outer+eye_left_inner))).astype(int)
    D_en = eye_ave_height-nose_height
    D_nm = nose_height-mouth_height


    for i in range(len(faces_list)):

        if (eye_right_inner[i] != 0) and (mouth_height[i] != 0):
            cheek_right = cv2.Sobel(gray_faces_list[i][int(cheeks_right_upper[i]):int(cheeks_right_lower[i]),int(cheeks_right_inner[i]):int(cheeks_right_outer[i])],-1,1,1)
            cheek_right_b = cv2.threshold(cheek_right,40,1,cv2.THRESH_BINARY)[1]
            cheek_right_pa = cheek_right_b.shape[0]*cheek_right_b.shape[1]
            cheek_right_wa = np.sum(cheek_right_b)
            cheek_right_pa_sum = np.sum(cheek_right)
            cheek_right_wa_sum = np.sum(cheek_right*cheek_right_b)
            cheek_right_density[i] = cheek_right_wa/cheek_right_pa
            if cheek_right_wa != 0:
                cheek_right_depth[i] = cheek_right_wa_sum/(cheek_right_wa*256)
            else:
                cheek_right_depth[i] = 0
            cheek_right_var[i] = cheek_right_pa_sum/(cheek_right_pa*256)

        if (eye_left_inner[i] != 0) and (mouth_height[i] != 0):
            cheek_left = cv2.Sobel(gray_faces_list[i][int(cheeks_left_upper[i]):int(cheeks_left_lower[i]),int(cheeks_left_outer[i]):int(cheeks_left_inner[i])],-1,1,1)
            ret,cheek_left_b = cv2.threshold(cheek_left,40,1,cv2.THRESH_BINARY)
            cheek_left_pa = cheek_left_b.shape[0]*cheek_left_b.shape[1]
            cheek_left_wa = np.sum(cheek_left_b)
            cheek_left_pa_sum = np.sum(cheek_left)
            cheek_left_wa_sum = np.sum(cheek_left*cheek_left_b)
            cheek_left_density[i] = cheek_left_wa/cheek_right_pa
            if cheek_left_wa != 0:
                cheek_left_depth[i] = cheek_left_wa_sum/(cheek_left_wa*256)
            else:
                cheek_left_depth[i] = 0
            cheek_left_var[i] = cheek_left_pa_sum/(cheek_left_pa*256)

        if (eye_right_inner[i] != 0):
            eye_corners_right = cv2.Sobel(gray_faces_list[i][int(eye_corners_right_upper[i]):int(eye_corners_right_lower[i]),int(eye_corners_right_inner[i]):int(eye_corners_right_outer[i])],-1,1,1)
            ret,eye_corners_right_b = cv2.threshold(eye_corners_right,40,1,cv2.THRESH_BINARY)
            eye_corners_right_pa = eye_corners_right_b.shape[0]*eye_corners_right_b.shape[1]
            eye_corners_right_wa = np.sum(eye_corners_right_b)
            eye_corners_right_pa_sum = np.sum(eye_corners_right)
            eye_corners_right_wa_sum = np.sum(eye_corners_right*eye_corners_right_b)
            eye_corners_right_density[i] = eye_corners_right_wa/eye_corners_right_pa
            if eye_corners_right_wa != 0:
                eye_corners_right_depth[i] = eye_corners_right_wa_sum/(eye_corners_right_wa*256)
            else:
                eye_corners_right_depth[i] = 0
            eye_corners_right_var[i] = eye_corners_right_pa_sum/(eye_corners_right_pa*256)

        if (eye_left_inner[i] != 0):
            eye_corners_left = cv2.Sobel(gray_faces_list[i][int(eye_corners_right_upper[i]):int(eye_corners_right_lower[i]),int(eye_corners_left_outer[i]):int(eye_corners_left_inner[i])],-1,1,1)
            ret,eye_corners_left_b = cv2.threshold(eye_corners_left,40,1,cv2.THRESH_BINARY)
            eye_corners_left_pa = eye_corners_left_b.shape[0]*eye_corners_left_b.shape[1]
            eye_corners_left_wa = np.sum(eye_corners_left_b)
            eye_corners_left_pa_sum = np.sum(eye_corners_left)
            eye_corners_left_wa_sum = np.sum(eye_corners_left*eye_corners_left_b)
            eye_corners_left_density[i] = eye_corners_left_wa/eye_corners_left_pa
            if eye_corners_left_wa != 0:
                eye_corners_left_depth[i] = eye_corners_left_wa_sum/(eye_corners_left_wa*256)
            else:
                eye_corners_left_depth[i] = 0
            eye_corners_left_var[i] = eye_corners_left_pa_sum/(eye_corners_left_pa*256)

        if (eye_right_inner[i] != 0) and (eye_left_inner[i] != 0):
            brow = cv2.Sobel(gray_faces_list[i][int(brow_upper[i]):int(brow_lower[i]),int(brow_left[i]):int(brow_right[i])],-1,1,1)
            ret,brow_b = cv2.threshold(brow,40,1,cv2.THRESH_BINARY)
            brow_pa = brow_b.shape[0]*brow_b.shape[1]
            brow_wa = np.sum(brow_b)
            brow_pa_sum = np.sum(brow)
            brow_wa_sum = np.sum(brow*brow_b)
            brow_density[i] = brow_wa/brow_pa
            if brow_wa != 0:
                brow_depth[i] = brow_wa_sum/(brow_wa*256)
            else:
                brow_depth[i] = 0
            brow_var[i] = brow_pa_sum/(brow_pa*256)

        if (eye_right_inner[i] != 0) and (eye_left_inner[i] != 0) and (mouth_height[i] != 0):
            R_em[i] = abs(D_em[i]/D_ee[i])

        if (eye_right_inner[i] != 0) and (eye_left_inner[i] != 0) and (mouth_height[i] != 0) and (nose_height[i] != 0):
            R_enm[i] = abs(D_en[i]/D_nm[i])


        face_features[i,0] = cheek_right_density[i]
        face_features[i,1] = cheek_right_depth[i]
        face_features[i,2] = cheek_right_var[i]
        face_features[i,3] = cheek_left_density[i]
        face_features[i,4] = cheek_left_depth[i]
        face_features[i,5] = cheek_left_var[i]
        face_features[i,6] = eye_corners_right_density[i]
        face_features[i,7] = eye_corners_right_depth[i]
        face_features[i,8] = eye_corners_right_var[i]
        face_features[i,9] = eye_corners_left_density[i]
        face_features[i,10] = eye_corners_left_depth[i]
        face_features[i,11] = eye_corners_left_var[i]
        face_features[i,12] = brow_density[i]
        face_features[i,13] = brow_depth[i]
        face_features[i,14] = brow_var[i]
        face_features[i,15] = R_em[i]
        face_features[i,16] = R_enm[i]
     
featurecompute(trainsetC, trainsetC.copy(),trainset)
featurecompute(testsetC, testsetC.copy(), testset)
t3 = time.time
print('Timeused:',t3-t2)
print('finished')
```
